```python
import pandas as pd

import tensorflow as tf
import utils
import numpy as np
import matplotlib.pyplot as plt
from Model import Model
%matplotlib inline

from transformers import pipeline
import gradio as gr
```

```python
dfs = []

dfs.append(pd.read_csv('game_1-500_min7.csv'))
dfs.append(pd.read_csv('game_501-1000_min7.csv'))
dfs.append(pd.read_csv('game_1001-1500_min7.csv'))
dfs.append(pd.read_csv('game_1501-2000_min7.csv'))
dfs.append(pd.read_csv('game_2001-3000_min7.csv'))
dfs.append(pd.read_csv('game_3001-4000_min7.csv'))
dfs.append(pd.read_csv('game_4001-5000_min7.csv'))
dfs.append(pd.read_csv('game_5001-10000_min7.csv'))
# dfs.append(pd.read_csv('game_5501-6000_min7.csv'))

df = pd.concat(dfs)
df
```

```python
df = utils.clean_data(df)
df = utils.remove_duplicates(df)
pivot_df = utils.pivot_with_threshold(df, 8)
normalized_df = utils.normalize(pivot_df)
```
