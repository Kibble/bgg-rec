```python
# Bohnanza, CATAN (46), Diplomacy, Go, Magic: The Gathering, RoboRally, 
def predict(catan, diplomacy, go, mtg):
    col_list = normalized_df.columns.to_list()
    input = [0] * len(normalized_df.columns)
    # Catan
    input[col_list.index('CATAN')] = float(catan)/10.0
    input[col_list.index('Diplomacy')] = float(diplomacy)/10.0
    input[col_list.index('Go')] = float(go)/10.0
    input[col_list.index('Magic: The Gathering')] = float(mtg)/10.0
    pred = model.predict(input, normalized_df.columns, 10)
    pred = pred.loc[pred['RecommendationScore'] > 2].copy()
    pred = pred[:10]
    return pred
```

```python

theme = gr.themes.Base(
    primary_hue="pink",
    secondary_hue="fuchsia",
    neutral_hue="pink"
)

theme.set(slider_color='#EC4899')

app = gr.Interface(
    fn=predict,
    title="Board Game Recommendation Engine",
    theme=theme,
    # theme='monochrome',
    # theme=gr.Base(
    #     primary_hue='pink',
    #     secondary_hue='yellow'
    # ),
    inputs=[
        gr.Slider(1, 10, value=5, step=1, label="Catan"),
        gr.Slider(1, 10, value=5, step=1, label="Diplomacy"),
        gr.Slider(1, 10, value=5, step=1, label="Go"),
        gr.Slider(1, 10, value=5, step=1, label="Magic: The Gathering")
    ],
    outputs=gr.DataFrame()
    # outputs=gr.Textbox(lines=10, label="Answer", show_copy_button=True)
)

# Launch the app.
# app.launch(show_error=True, share=True)
app.launch(show_error=True)
```
